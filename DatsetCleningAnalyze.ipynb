```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("dataset.xls")
```

```python
# Первичный анализ
# Упоппусков, у которых degree > 2 (Phd, ;)
```

```python
def count_straight_stress(row):
    sm = row["source_6"] + row["source_5"] + row["source_4"] + row["source_3"] + row["source_2"] + row["source_1"]    
    if not sm is None:
        return sm
    else:
        return 0
```

```python
df["straight_stress"] = df.apply (lambda row: count_straight_stress(row), axis=1) # общий стресс сурсов
```

```python
for i in range(1, 9):
    sns.lineplot(data=df, x="stress", y=f"source_{i}")
    plt.show()

plt.legend([str(i) for i in range(1, 9)])
```
