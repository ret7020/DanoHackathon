```python
def cast_get(x):
    study = x.week_1 + x.week_2
    work = x.week_3
    mod = abs(study - work)
    
    if work < study and mod >= 0.1 * study:
        return "study"
    elif work > study and mod >= 0.1 * study:
        return "work"
    else:
        return "complex"
```

```python
df["cast"] = df.apply (lambda row: cast_get(row), axis=1)
```

```python
data = df.value_counts("cast")
labels = ['study', 'complex', "work"]
colors = sns.color_palette('hls')[0:3]
plt.pie(data, labels = labels, colors = colors, autopct='%.0f%%')
plt.show()
```
