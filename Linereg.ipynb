```python
import sklearn
import pandas as pd
```

```python
df = pd.read_excel("dataset.xls")
```

```python
df = df.dropna()
```

```python
df
```

```python
from sklearn.linear_model import LinearRegression
sklearn.linear_model.LinearRegression()
```

```python
# Input (X):
# degree, semester, gpa
# Ouput (Y):
# source_6
X = df[["degree", "semester", "gpa"]]
y = df[["source_6"]]
```

```python
y
```

```python
regressor = LinearRegression()
regressor.fit(X, y)
```

```python
y_pred = regressor.predict(X)
```

```python
y_pred
```

```python
X["semester"]
```

```python
import matplotlib.pyplot as plt
plt.scatter(X["semester"], y, color = 'red')
plt.plot(X["semester"], regressor.predict(X), color = 'blue')
plt.title('Semester to stress')
plt.xlabel('degree, semester, gpa')
plt.ylabel('stress')
plt.show()
```
